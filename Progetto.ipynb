```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
```

```python
# Percorso della cartella contenente le immagini
image_folder = 'C:\\Users\\matte\\OneDrive\\Desktop\\Machine Learning\\Progetto\\immagini\\1c'

# Elenco di tutti i file di immagine
images = [img for img in os.listdir(image_folder) if img.endswith(('png', 'jpg', 'jpeg'))]
images.sort()  # Opzionale, ma utile per mantenere un ordine costante
```

```python
# Lista per salvare i dati
data = []

# Loop per visualizzare ogni immagine e automaticamente assegnare l'etichetta '1c'
for image in images:
    # Mostra l'immagine
    img_path = os.path.join(image_folder, image)
    img = plt.imread(img_path)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    
    # Imposta automaticamente l'etichetta '1c' per ogni immagine
    label = '1c'
    
    # Salva i risultati in una lista
    data.append([image, label])
    
    # Pulisci l'output per la prossima immagine
    clear_output(wait=True)

print("Etichettatura completata.")
```

Caricare e Ridimensionare le Immagini: Ridimensioneremo tutte le immagini a una dimensione standard, ad esempio 64x64 pixel.
Normalizzare i Pixel: Convertire i valori dei pixel da 0-255 a 0-1.
Data Augmentation: Applicare trasformazioni come rotazioni, spostamenti e zoom per aumentare la varietà nel training set e migliorare la generalizzazione del modello.

```python
from PIL import Image, ImageOps
import os
import numpy as np
import matplotlib.pyplot as plt

def load_and_preprocess_image(image_path, target_size=(64, 64)):
    img = Image.open(image_path)
    img = img.resize(target_size)
    img_array = np.array(img) / 255.0
    return img_array

def augment_image(img):
    # Applica una rotazione
    rotated = img.rotate(30)  # Ruota di 30 gradi
    
    # Applica un flip orizzontale
    flipped = ImageOps.mirror(img)
    
    return rotated, flipped

# Percorso della cartella delle immagini
folder_path = 'C:\\Users\\matte\\OneDrive\\Desktop\\Machine Learning\\Progetto\\Image'
images = []

# Carica e mostra alcune immagini trasformate
for file_name in os.listdir(folder_path):
    if file_name.endswith(('.png', '.jpg', '.jpeg')):
        image_path = os.path.join(folder_path, file_name)
        original_image = Image.open(image_path)
        rotated_image, flipped_image = augment_image(original_image)

        plt.figure(figsize=(12, 4))
        plt.subplot(1, 3, 1)
        plt.imshow(np.array(original_image), cmap='gray')
        plt.title('Original')
        plt.axis('off')
        
        plt.subplot(1, 3, 2)
        plt.imshow(np.array(rotated_image), cmap='gray')
        plt.title('Rotated')
        plt.axis('off')

        plt.subplot(1, 3, 3)
        plt.imshow(np.array(flipped_image), cmap='gray')
        plt.title('Flipped')
        plt.axis('off')

        plt.show()

        break  # Mostriamo solo la prima immagine per esempio
```
